```python
#Base class 1
class Animal:
  def __init__(self,name):
    self.name = name

  def speak(self):
    print("Subclasses must implement this method")


#Base class 2
class Pet:
  def __init__(self, owner):
    self.owner = owner


#Dervied class
class Dog(Animal , Pet):
  def __init__(self,name,owner):
    Animal.__init__(self,name)
    Pet.__init__(self,owner)

  def speak(self):
    return f"{self.name} say woof"


dog = Dog("Buddy" , "Siddharth")
print(dog.speak())
print(f"Owner:{dog.owner}")
```

```python
#Base class
class Animal:
  def speak(self):
    return "Sound of the animal"


#Derived class
class Dog(Animal):
  def speak(self):
    return "Woof!"

class Cat(Animal):
  def speak(self):
    return "Meow1"

#Function that demonstrates Polymorphism
def animal_speak(animal):
  print(animal.speak())

dog = Dog()
cat = Cat()
print(dog.speak())
print(cat.speak())
animal_speak(dog)
```

```python
class shape:
  def area(self):
    return "the area of the figure"

#Derived class
class Rectangle(shape):
  def __init__(self, width , height):
    self.width = width
    self.height = height

  def area(self):
    return self.width * self.height

  ##Derived class 2

class Circle(shape):

  def __init__(self,radius):
    self.radius = radius

  def area(self):
    return 3.14*self.radius

  ## Function that demonstrates polymorphism

def print_area(shape):
  print(f"the area is {shape.area()}")

rectangle = Rectangle(4,5)
circle = Circle(3)

print_area(rectangle)
print_area(circle)
```

```python
#polymorphism with abstract base class
from abc import ABC, abstractmethod

## Define an abstract class
class Vehicle(ABC):
  @abstractmethod
  def start_engine(self):
    pass

## Derived class
class Car(Vehicle):
  def start_engine(self):
    return "Car engine started"

## Derived class 2
class Motorcycle(Vehicle):
  def start_engine(self):
    return "Motorcycle engine started"

## functionn to demonstrate polymorphism
def start_vehicle(Vehicle):
  print(Vehicle.start_engine())

## create objects of car and motorcycle

car = Car()
motorcycle = Motorcycle()

start_vehicle(car)
```

```python
## Abstraction
from abc import ABC, abstractmethod

#Abstract class
class Vehicle(ABC):
  def drive(self):
    print("the vehicle is used for driving")


  @abstractmethod
  def start_engine(self):
    pass

class Car(Vehicle):
  def start_engine(self):
    print("car engine Started")

def operate_vehicle(vehicle):
  vehicle.start_engine()
  vehicle.drive()

car = Car()
operate_vehicle(car)
```

```python
## Magic methods

class Person:
  def __init__(self,name,age):
    self.name = name
    self.age = age

  def __str__(self):
    return f"{self.name},{self.age} years old"

  def __repr__(self):
    return f"Person(name = {self.name},age = {self.age})"

person = Person("Siddharth", 21)
print(person)
print(repr(person))
```

```python
### Comman Operatror Overloading Magic Methods

'''
__add__(self, other): Adds two objects   using the + operator
__sub__(self, other): Subtracts  two objects using the _ operator
__nul__(self, other): Multiplies two objects using the + operator
__truediv__(self, other): Divides two objects are equal using the / operator
__eq__(seld, other): Checks if two objects are equal using the ++ operator.
__lt__(self, other) : Checks if two objects is less than another using the < operator

'''
```

```python
class Vector:
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def __add__(self, other):
    return Vector(self.x + other.x, self.y + other.y)

  def __sub__(self, other):
    return Vector(self.x - other.x, self.y - other.y)

  def __mul__(self, other):
    return Vector(self.x * other.x , self.y * other.y)

  def __eq__(self, other):
    return self.x == other.x and self.y == other.y

  def __repr__(self):
    return f"Vector({self.x} , {self.y})"


## create objects of the vector Class
v1 = Vector(2,3)
v2 = Vector(4,5)

print(v1 + v2)
```
