### Naive Bayes Classifiers

Naive Bayes classifiers are a family of simple probabilistic classifiers based on applying Bayes' theorem with strong (naive) independence assumptions between the features. They are known for their simplicity, efficiency, and effectiveness, especially in text classification and other high-dimensional datasets.

There are several types of Naive Bayes classifiers, each suited for different types of data distribution:

1.  **Gaussian Naive Bayes**: Used when features are continuous and assumed to be distributed according to a Gaussian (normal) distribution.
2.  **Multinomial Naive Bayes**: Suitable for discrete features (e.g., word counts for text classification). It is commonly used for document classification problems.
3.  **Bernoulli Naive Bayes**: Designed for binary or boolean features. It assumes that each feature is a Bernoulli distributed variable (i.e., it can take on only two values, typically 0 or 1).

#### Gaussian Naive Bayes (Recap)

As already implemented, Gaussian Naive Bayes assumes that continuous features associated with each class are distributed according to a Gaussian distribution. This is often a good default choice for continuous data.

#### Evaluating the Gaussian Naive Bayes Model

Let's add more evaluation metrics to assess the performance of our Gaussian Naive Bayes model, such as a Confusion Matrix and a Classification Report.

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred_gnb = model.predict(X_test_imputed)

print("--- Gaussian Naive Bayes Evaluation ---")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_gnb))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_gnb))
```

#### Multinomial Naive Bayes

Multinomial Naive Bayes is suitable for classification with discrete features (e.g., word counts in text classification). The feature values should be counts or frequencies. While our current features (`Pclass`, `Age`, `Fare`) are not strictly counts, `Pclass` could be treated as discrete, and `Age`/`Fare` could be binned or scaled for MultinomialNB if required. For demonstration, we'll apply it directly here, but ideally, preprocessing (like scaling or binning for continuous features) might be needed for optimal performance.

For `MultinomialNB`, feature values must be non-negative. We'll ensure this by replacing any negative values with 0 if they exist (though unlikely in this dataset).

```python
from sklearn.naive_bayes import MultinomialNB

# Ensure non-negative features for MultinomialNB
# Age and Fare can sometimes have outliers or be problematic if not handled,
# but for simplicity, we'll assume they are fine after imputation for now.
# If there were negative Fares (which shouldn't happen), we'd need to address it.
# Let's create a copy of inputs for MultinomialNB to be safe with potential future modifications
inputs_mnb = inputs.copy()

# A simple check to ensure no negative values before fitting MultinomialNB
# While unlikely for Age/Fare, it's good practice.
# We will fill NaN in Age with the mean to make it compatible.
inputs_mnb['Age'] = inputs_mnb['Age'].fillna(inputs_mnb['Age'].mean())

# Convert floats to integers if appropriate, or ensure non-negativity for MNB
# Pclass, female, male are already good.
# Age and Fare need to be non-negative. Also, MNB often expects count-like data.
# For demonstration, we'll just ensure non-negativity.
inputs_mnb[inputs_mnb < 0] = 0


X_train_mnb, X_test_mnb, y_train_mnb, y_test_mnb = train_test_split(inputs_mnb, target, test_size=0.2, random_state=42)

model_mnb = MultinomialNB()
model_mnb.fit(X_train_mnb, y_train_mnb)

print("--- Multinomial Naive Bayes Evaluation ---")
print(f"Multinomial Naive Bayes Accuracy: {model_mnb.score(X_test_mnb, y_test_mnb)}")

y_pred_mnb = model_mnb.predict(X_test_mnb)
print("Confusion Matrix:")
print(confusion_matrix(y_test_mnb, y_pred_mnb))
print("\nClassification Report:")
print(classification_report(y_test_mnb, y_pred_mnb))
```

#### Bernoulli Naive Bayes

Bernoulli Naive Bayes is designed for binary/boolean features. It explicitly models the presence or absence of a particular feature. It's often used when features represent whether a term is present or not in a document (as opposed to its count).

For our dataset, we can binarize `Age` and `Fare` (e.g., above/below mean/median) or use the existing binary `female`/`male` and discrete `Pclass` features directly. For `BernoulliNB`, features are typically 0 or 1. Let's create a binarized version of `Age` and `Fare` for this model, and use `Pclass` as is (it will be treated as binary if it only had 0s and 1s, otherwise it's just a numerical feature where its presence/absence is modeled).

Here, we'll simplify by thresholding `Age` and `Fare` at their means to create binary features.

```python
from sklearn.naive_bayes import BernoulliNB

# Create a binarized version of the inputs for BernoulliNB
inputs_bnb = inputs.copy()

# Fill NaN in Age with the mean for binarization
inputs_bnb['Age'] = inputs_bnb['Age'].fillna(inputs_bnb['Age'].mean())

# Binarize Age and Fare
inputs_bnb['Age_Binarized'] = (inputs_bnb['Age'] > inputs_bnb['Age'].mean()).astype(int)
inputs_bnb['Fare_Binarized'] = (inputs_bnb['Fare'] > inputs_bnb['Fare'].mean()).astype(int)

# Drop original Age and Fare columns, keep Pclass, female, male, and new binarized features
inputs_bnb = inputs_bnb.drop(columns=['Age', 'Fare'])

X_train_bnb, X_test_bnb, y_train_bnb, y_test_bnb = train_test_split(inputs_bnb, target, test_size=0.2, random_state=42)

model_bnb = BernoulliNB()
model_bnb.fit(X_train_bnb, y_train_bnb)

print("--- Bernoulli Naive Bayes Evaluation ---")
print(f"Bernoulli Naive Bayes Accuracy: {model_bnb.score(X_test_bnb, y_test_bnb)}")

y_pred_bnb = model_bnb.predict(X_test_bnb)
print("Confusion Matrix:")
print(confusion_matrix(y_test_bnb, y_pred_bnb))
print("\nClassification Report:")
print(classification_report(y_test_bnb, y_pred_bnb))
```
