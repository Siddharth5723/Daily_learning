```python
#Linear Regression -- Single variable
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data  = {
    'area': [2600, 3000, 3200, 3600, 4000, 4200, 4500, 4800, 5000, 5500],
    'price': [550000, 565000, 610000, 680000, 725000, 750000, 790000, 825000, 860000, 940000]
}

df = pd.DataFrame(data)
print(df)
```

```python
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area , df.price , color = "red" , marker ="+")
plt.show()
plt.tight_layout()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]], df.price)
```

```python
print(reg.predict([[3300]]))
```

```python
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area , df.price , color = "red" , marker ="+")
plt.plot(df.area , reg.predict(df[["area"]]) , color="blue")
plt.show()
plt.tight_layout()
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
data_1 = {
    'area': [2600, 3000, 3200, 3600, 4000, 4100, 4500, 4800, 5100, 5500],
    'bedrooms': [3, 4, 3, None, 5, 6, 5, 4, 5, 6],
    'age': [20, 15, 18, 30, 8, 8, 12, 5, 2, 10],
    'price': [550000, 565000, 610000, 595000, 760000, 810000, 790000, 850000, 920000, 940000]
}

df_2 = pd.DataFrame(data_1)
print(df_2)
```

```python
print(df_2)
```

```python
import math
median_bedrooms = math.floor(df_2.bedrooms.median())
median_bedrooms
```

```python
df_2.bedrooms = df_2.bedrooms.fillna(median_bedrooms)
print(df_2)
```

```python
reg1 = linear_model.LinearRegression()
reg1.fit(df_2[['area' , 'bedrooms' , 'age']] , df_2.price)
```

```python
reg1.predict([[3000 , 3 , 40]])
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot: Area, Age, and Price
ax.scatter(df_2['area'], df_2['age'], df_2['price'], c='purple', marker='o', s=100)

ax.set_xlabel('Area (sq ft)')
ax.set_ylabel('Age (Years)')
ax.set_zlabel('Price ($)')
ax.set_title('3D View: Area & Age vs Price')
plt.tight_layout()
plt.grid()
plt.show()
```
